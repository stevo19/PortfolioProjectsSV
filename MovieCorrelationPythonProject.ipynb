```python
#Import libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use ('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #this is to Adjust the configuration of the plots that we'll create
```

```python
#Read in Kaggle Movies Data 
df = pd.read_csv(r'C:\Users\stevenson\Desktop\movies.csv')
```

```python
#Let's explore the data
df.head()
```

```python
#Look for missing columns using a for loop
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Data types for our columns
df.dtypes
```

```python
#change the data type of the budget and gross columns to Integers
df= df.fillna(0)

df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
#df
```

```python
#order by gross revenue
df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
#if we want to see see the entire data
pd.set_option('display.max_rows', None)
```

```python
#checking for any null values
df.isnull().sum()
```

```python
#drop duplicated data for further cleaning 
df.drop_duplicates()
```

```python
df.head()
```

```python
#Let's get insights on what makes a movies sucessfull.
#Just by looking at the data without any visualization. Budget and the company that produced the movies would be my top 2 factors.
```

```python
#scatter plot visual budget vs gross
#---- Expectation-- As budget increases, gross revenue also increase
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Total Budget')
plt.show()

#---result- visualy we can see how the x and y are related based on the trend  of the scatter plot
```

```python
#Let's create a regression plot for clearer analysis using SEABORN 
sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color": "black"}, line_kws = {"color":"blue"} )
```

```python
df.dtypes
```

```python
correlation_matrix = df.corr(numeric_only = True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Let's numerized the columns for further exploration- given the columns numeric representation - for example US = 55, UK = 56
df_numerized = df 
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
#we can tell that the trend is upward but by how much?
#Let's look at a correlation table
#As we know there are different correlation methods. #Pearson, #Kendall, #Spearman

df.corr(numeric_only = True)
#df.corr(method= 'pearson')
```

```python
#bright colors = high corr
#darker colors = Low corr

correlation_matrix = df.corr(numeric_only = True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#those are just exploration--- basically changing the columns str to an assigned numeric value to see if there a correlation.
#for example: VOTE VS GROSS, BUDGET VS SCORE, GROSS VS RUNTIME
#As previously stated above the goal is to figure out what factors makes a movie succesfull and this correlation table shows that
```

```python
#Let's filter this down a bit 
correlation_mat = df.corr(numeric_only = False)
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#Let's Paired up the factors 
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#rank by the order of correlation
high_corr = sorted_pairs [(sorted_pairs) > 0.5]
high_corr
```
